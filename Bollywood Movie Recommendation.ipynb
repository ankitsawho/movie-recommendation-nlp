```python
myData = df.loc[(df['Origin/Ethnicity']=='Bollywood')]
newdf = pd.DataFrame(myData)
```

```python
newdf = newdf[["Title", "Plot"]]
newdf = newdf.set_index('Title')
```

```python
# Preprocessing
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('stopwords')

from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(text):
    text = text.lower()
    words = nltk.word_tokenize(text)
    my_sent = [lemmatizer.lemmatize(word) for word in words if word not in stopwords.words('english')]
    final_sent = ' '.join(my_sent)
    
    final_sent = final_sent.replace("n't", "not")
    final_sent = final_sent.replace("'m'", "am")    
    final_sent = final_sent.replace("'s'", "is")    
    final_sent = final_sent.replace("'re'", "are")    
    final_sent = final_sent.replace("'ll'", "will")    
    final_sent = final_sent.replace("'ve'", "have")
    final_sent = final_sent.replace("'d'", "would")
    return final_sent
```

```python
newdf['new_plot'] = newdf['Plot'].apply(preprocess) 
```

```python
newdf
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_movies = tfidf.fit_transform(newdf['new_plot'])

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(tfidf_movies, tfidf_movies)
```

```python
indices = pd.Series(newdf.index)
```

```python
def recommend(title, cosine_similarity=similarity):
    try:
        index = indices[indices==title].index[0]
        similarity_scores = pd.Series(cosine_similarity[index]).sort_values(ascending=False)
        top_10_movies = list(similarity_scores.iloc[1:11].index)
        recommend_movies = [list(newdf.index)[i] for i in top_10_movies]
        return recommend_movies
    except:
        print('No movies found')
```

```python
to_csv_data = newdf.drop('new_plot', axis=1)
```

```python
to_csv_data.to_csv('bollywood.csv', encoding='utf-8')
```
